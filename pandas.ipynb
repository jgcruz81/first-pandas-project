This project is my first exploratory analysis of a dataset with jupyter notebooks. I am using a dataset  of used cars from eBay Kleinanzeigen that was originally uploaded to Kaggle.

```python
import pandas as pd
import re
```

```python
autos = pd.read_csv("autos.csv", encoding = "Latin-1")
autos.info()
autos.head()
```

We can see from the above methods that we have some null values as well as camelcase for column names. In the cell below I changed a couple of the column names and imported a method to covert camelcase to snakecase, which as you can probably guess will convert the column names to snakecase.

```python
columns = autos.columns
autos.rename({"yearOfRegistration": "registration_year"}, axis=1, inplace=True)
autos.rename({"monthofRegistration": "registratioon_month"}, axis=1, inplace=True)
autos.rename({"notRepairedDamage": "unrepaired_damage"}, axis=1, inplace=True)
autos.rename({"dateCreated": "ad_created"}, axis=1, inplace=True)
autos.rename({"nrOfPictures": "num_photos"}, axis=1, inplace=True)


def camel_to_snake(column_name):
  name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', column_name)
  return re.sub('([a-z0-9])([A-Z])', r'\1_\2', column_name).lower()

new_columns = []
for c in autos.columns:
    new_columns.append(camel_to_snake(c))
autos.columns = new_columns
autos.head()
```

```python
autos.describe(include="all")
```

From the above cell we can see that there are some values that are not consistent with standard practices. By that I mean the 'price' and 'odometer' columns have string values instead of integers, which explains why the mean was not able to be calculated. I will proceed to clean up the data by converting certain columns to the integer data type in order to make the data easier to analyze. I will also rename the 'odomter' column to 'odomter_km' in order to not lose any information.

```python
autos["price"] = (autos["price"]
                          .str.replace("$","")
                          .str.replace(",","")
                          .astype(int)
                          )

autos["odometer"] = (autos["odometer"]
                             .str.replace("km","")
                             .str.replace(",","")
                             .astype(int)
                             )
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos.head()
```

Since the number of photos for this dataset is 0 for all rows, the 'photos' column is a candidate for being dropped because it contains redudant information

```python
autos = autos.drop("num_photos", axis=1)
```

##Brand

```python
autos["brand"].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

From the cells above we can see that from all the brands available that Audi, Mercedes-Benz, and BMW are on average more expensive that the other popular brands. With the most common vehicle brand, Volkswagen, residing around on average between the luxury brands and the economical ones. Next let's look at mileage to find a correlation between the 'price' and 'odometer' columns.

```python
brand_mean_odometer = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_odometer = brand_only["odometer_km"].mean()
    brand_mean_odometer[brand] = int(mean_odometer)

brand_mean_odometer
```

```python
mean_mileage = pd.Series(brand_mean_odometer).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_mileage,columns=['mean_odometer'])
brand_info["mean_price"] = mean_prices
brand_info
```
